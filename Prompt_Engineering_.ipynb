```python
def get_response(prompt):
  response = openai.chat.completions.create(
        model = 'gpt-4o-mini',
        messages = [{ "role": "user", "content": prompt }],
    )
  return IPython.display.Markdown(response.choices[0].message.content)
```

```python
get_response('After Rain, the roads are')
```

**Task 1.1 Text Summarization**

```python
def get_response_with_params(prompt,temperature=0.5,max_tokens=500):
  response = openai.chat.completions.create(
        model = 'gpt-4o-mini',
        messages = [{ "role": "user", "content": prompt }],
        temperature=temperature,
        max_tokens=max_tokens
    )
  return IPython.display.Markdown(response.choices[0].message.content)
```

```python
get_response_with_params('After Rain, the roads are')
```

```python
prompt = """Artificial Intelligence (AI) has become one of the most transformative technologies of the modern era. It enables machines to perform tasks that traditionally required human intelligence, such as learning, reasoning, decision-making, and language understanding. AI systems are widely used in various domains including healthcare, finance, transportation, and education. For example, in healthcare, AI helps doctors analyze medical images, predict diseases, and personalize treatment plans. In finance, AI is used for fraud detection, algorithmic trading, and risk assessment. Despite its numerous benefits, AI also raises important ethical concerns such as data privacy, bias in algorithms, and the potential impact on employment. As AI technologies continue to evolve, it is crucial for researchers, policymakers, and organizations to ensure that these systems are developed and used responsibly for the benefit of society.

Explain the above in three sentence:"""

get_response_with_params(prompt)
```

**Task 1.2 Instructional Prompting**

```python
prompt = """
Explain Random Forest algorithm.

Format your answer as:
1. Definition
2. How the algorithm works
3. Advantages
4. Limitations
"""
get_response_with_params(prompt)
```

**Task 1.3 Text Classification**

```python
prompt = """Classify the text into neutral, negative or positive.

Text: I think the food was okay.

Sentiment:"""

get_response_with_params(prompt)
```

```python
prompt = """
You are a university professor teaching data science.
Explain Overfitting to undergraduate students with a simple example.
"""
get_response_with_params(prompt)
```
