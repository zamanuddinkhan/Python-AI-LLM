**Chain-of-Thought (CoT) Prompting**

Concept: The model is instructed to think through problems step-by-step, explicitly writing out intermediate reasoning before the final answer. This often improves accuracy on complex reasoning or math tasks.

```python
!pip install groq
```

```python
from google.colab import userdata
GROQ_API_KEY=userdata.get('GROQ_API_KEY')
import IPython
```

```python
#Example 1

prompt = """
You are a reasoning assistant. Solve the following problem step by step,
and then give the final answer.

Problem: If a train travels 60 km in 1.5 hours, what is its average speed?
Let's reason step by step:
"""

#Note: The phrase “Let’s think step by step” is a known CoT trigger that makes GPT models reason more systematically.
```

```python
from groq import Groq
client = Groq(api_key=userdata.get('GROQ_API_KEY'))

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",  # or "gpt-4-turbo"
    messages=[
        {"role": "system", "content": "You are a helpful math tutor."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.2,
)
IPython.display.Markdown(response.choices[0].message.content)
```

**Tree-of-Thoughts**

ToT expands CoT by exploring multiple reasoning paths (branches) before deciding on the best answer.

You can structure your prompt to explicitly tell the model to generate, evaluate, and choose from multiple reasoning paths.

Example: Logic Puzzle

“A farmer has chickens and cows. There are 30 heads and 74 legs in total.
How many chickens and cows are there?”

```python
prompt = """
You are a reasoning assistant solving a logic puzzle.

Problem:
A farmer has chickens and cows. There are 30 heads and 74 legs in total.
How many chickens and cows are there?

Follow this process:
1. Generate at least two reasoning paths (labeled Path 1, Path 2, etc.)
2. Explore each path step by step.
3. Compare which path is logically consistent.
4. Give the final correct answer.

Let's begin.
"""
```

```python
chat_completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "You are an expert logical reasoner."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.3,
)

IPython.display.Markdown(response.choices[0].message.content)
```

**ToT with Multi-Round Evaluation**

With two prompts Generator Prompt and Evaluator Prompt

```python
# Step 1: Generate possible reasoning paths
generate_prompt = """
Generate 3 different reasoning paths for this problem:
A 12-year-old boy is twice as old as his sister. How old will the sister be when he is 30?
"""

gen_response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": generate_prompt}],
)
paths = gen_response.choices[0].message.content
print("Generated paths:\n", paths)

# Step 2: Evaluate which path is most logical
evaluate_prompt = f"""
Here are some reasoning paths for a math problem:
{paths}

Evaluate each one and decide which path gives the correct answer.
Return only the best reasoning and the final answer.
"""

eval_response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": evaluate_prompt}],
)
print("Best path:\n")
IPython.display.Markdown(eval_response.choices[0].message.content)
```

**Step-Back Prompting** (introduced by Google DeepMind, 2023) is a reasoning strategy that tells the model to:

“Step back” from the details of the question,
think about it in a more abstract or general way first,
then use that higher-level insight to answer the specific question.

It’s basically two stages:

- Abstraction: Generate a general principle or concept behind the question.

- Application: Use that abstract reasoning to give the final answer.

- Normal Prompt:

Why does a metal rod expand when heated?

LLMs might answer:
Because heating increases particle movement, causing expansion.

- Step-Back Prompt:

First, step back and describe the general principle behind what happens when materials are heated.
Then, apply that principle to explain why a metal rod expands when heated.

- Better Output:

General principle: When materials are heated, their particles gain kinetic energy and move more vigorously, increasing the average distance between them.
Application: In metals, atoms vibrate more when heated, expanding the entire rod.

```python
prompt = """
Step-back prompting task:

Question: Why do darker surfaces get hotter in sunlight than lighter surfaces?

Instructions:
1. Step back and identify the general physical principle involved.
2. Then apply it specifically to darker vs. lighter surfaces.
3. Finally, summarize your answer concisely.
"""

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "You are a physics reasoning expert."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.3
)

IPython.display.Markdown(response.choices[0].message.content)
```

**Combining Step-Back with CoT (Hybrid Prompt)**

Prompt Template:

You are an expert reasoner.

Question: {question}

Instructions:
1. Step back and describe the general concept or principle related to this question.
2. Then apply that principle to the specific case.
3. Finally, give a clear, concise final answer.


```python
prompt = """
You are a scientific reasoning assistant.

Task: Answer this question using a step-back approach.

Question: Why does water boil faster at higher altitudes?

Follow these steps:
1. Step back to recall the general principle relating pressure and boiling point.
2. Explain how that principle applies to altitude.
3. Show your reasoning step by step before giving the final answer.
"""
```

```python
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "You are a physics reasoning expert."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.3
)

IPython.display.Markdown(response.choices[0].message.content)
```

**Self Consisteancy Prompts**

Self-Consistency Prompting is an advanced prompt engineering technique used to improve the reliability and accuracy of responses from Large Language Models (LLMs). Instead of generating a single reasoning path, the model is prompted to produce multiple independent reasoning paths for the same question. The final answer is then selected based on the most consistent or most frequent result among the generated solutions.

```python
prompt="""Solve the following problem step-by-step and provide the final answer.

A shop sells 18 notebooks each day.
How many notebooks will the shop sell in 25 days?

Explain your reasoning clearly and provide the final answer at the end."""
```

```python
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "You are a reasoning expert."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.3
)

IPython.display.Markdown(response.choices[0].message.content)
```
